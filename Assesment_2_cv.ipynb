**Question 1.** lot the k-NN decision boundaries of the IRIS dataset with two
features and three classes. Implement with Python for k=1 and k=5.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load the IRIS dataset and select the first two features
iris = datasets.load_iris()
data = iris.data[:, :2]
target = iris.target
```

```python
# Create an instance of the KNeighborsClassifier with k=1 and fit the data
knn_1 = KNeighborsClassifier(n_neighbors=1)
knn_1.fit(data, target)
```

```python
# Create an instance of the KNeighborsClassifier with k=5 and fit the data
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(data, target)
```

```python
# Prepare the meshgrid to plot the decision boundaries
h = 0.02  # Step size in the mesh
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
# Define the colormap for the plot
cmap = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
```

```python
# Plot the decision boundaries for k=1
plt.subplot(1, 2, 1)
Z = knn_1.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.8)
plt.scatter(data[:, 0], data[:, 1], c=target, cmap=cmap, edgecolor='k')
plt.title("k-NN Decision Boundaries (k=1)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
# Plot the decision boundaries for k=5
plt.subplot(1, 2, 2)
Z = knn_5.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.8)
plt.scatter(data[:, 0], data[:, 1], c=target, cmap=cmap, edgecolor='k')
plt.title("k-NN Decision Boundaries (k=5)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
